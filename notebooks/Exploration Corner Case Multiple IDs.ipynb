```python
df = d_train.data[["conference","conference_name","conference_year"]].drop_duplicates()
check_ids = df.groupby(["conference_name","conference_year"]).count().reset_index() \
    .merge(df,left_on=["conference_name","conference_year"],right_on=["conference_name","conference_year"]) \
    .sort_values(by=["conference_x","conference_name"],ascending=False)
```
